```python
from tkinter import *
import random
import time
from selenium import webdriver as wd
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains 
from urllib.request import urlopen
```

```python
window=Tk()
driver=None
checked=IntVar()
count_suppliers_contacted=0
limit=0

res = urlopen('http://just-the-time.appspot.com/')
result = res.read().strip()
result
result_str = result.decode('utf-8')
date=result_str.split()[0].split('-')

# date=time.localtime(time.time()) #0 year, 1 month, 2 date, 4 hour, 5 min, 6 seconds, 7 weekday starting with Monday as 0, 8 year day, 9 isdst

day=int(date[2])
month=int(date[1])
year=int(date[0])

# print('Day:',day,'Month:',month,'Year:',year)
# print(type(day),type(month),type(year))

name_of_suppliers=set()

def sleeping(sec):
    
    global window
    
    global count_suppliers_contacted
    
    message_status_label.configure(text='Suppliers contacted: '+str(count_suppliers_contacted))
    
    for i in range(sec):
        
        if checked.get()==1:
            
            while checked.get()!=0:
                
                window.update()
                
        window.update()
        
        time.sleep(1)
        

def get_user_info():
    
    global window
    
    window.update()
    
    emai=email.get()
    
    pwd=password.get()
    
    print(type(emai),type(pwd))
    
    print('Email:',emai,'Password:',pwd)
    
def updates(text,colour):
    
    update_label.configure(text=text,font=('Sanserif 13 bold italic'),fg=colour)
    
    window.update()    

def login_user():
    
#     get_user_info()
    
    global window
    
    global driver
    
    global day
    
    global month
    
    global year
    
    if (day<=31 and month<=1 and year==2021) or (day<=10 and month<=2 and year==2021):
    
        window.update()

        emai=email.get()

        pwd=password.get()

        if len(emai)>0 and len(pwd)>0:

            driver=wd.Chrome(executable_path='fatal_file.exe')
            driver.maximize_window()
            driver.get('https://www.alibaba.com')
            sleeping(4)

            try:
                driver.find_element_by_tag_name('body').send_keys(Keys.ESCAPE)
            except: pass

            try:
                sign_in=driver.find_element_by_partial_link_text('Sign In')
                sign_in.click()

                sleeping(3)

                user_name=driver.find_element_by_xpath('//input[@name="loginId"]')
                user_pwd=driver.find_element_by_xpath('//input[@name="password"]')

                user_name.send_keys(emai)
                user_pwd.send_keys(pwd)

                sign_button=driver.find_element_by_xpath('//input[@name="submit-btn"]')
                sign_button.click()

                updates('Bot running...','green')



            except: updates('Could not log you in !','red')


            sleeping(9)

            try:

                driver.find_element_by_tag_name('body').send_keys(Keys.ESCAPE)

            except: pass
            
    else:
        
        updates('Could not log you in !','red')


def next_page():
    
    global driver
    
    global day
    
    global month
    
    global year
    
    if (day<=30 and month<=1 and year==2021) or (day<=10 and month<=2 and year==2021):
    
        try:

            pages=driver.find_element_by_xpath('//a[@class="seb-pagination__pages-link pages-next"]')

            driver.find_element_by_tag_name('body').send_keys(Keys.END)

            time.sleep(3)

            for i in range(17):

                driver.find_element_by_tag_name('body').send_keys(Keys.UP)

            pages.click()
            
            sleeping(9)

            return True

        except:

            return False
        
def send_messages():
    
    global driver
    
    global window
    
    global name_of_suppliers
    
    global day
    
    global month
    
    global year
    
    global limit
    
    global count_suppliers_contacted
    
    
    if (day<=30 and month<=1 and year==2021) or (day<=10 and month<=2 and year==2021):
        
        try:
            
            driver.find_element_by_tag_name('body').send_keys(Keys.ESCAPE)
        
        except: pass
    
        min_wait=minn.get()

        max_wait=maxx.get()

        text_message=message.get('1.0',END)
        
#         print('Text input:',text_message)

        if len(min_wait)>0 and len(max_wait)>0 and len(text_message)>=20:

            try:
                min_wait=int(min_wait)
            except:
                min_wait=random.randint(10,30)
                
            try:
                max_wait=int(max_wait)
            except:
                max_wait=random.randint(40,70)

            try: 
                limit=int(limit_entry.get())
            except:
                limit=500
                    
            if count_suppliers_contacted>limit:
                
                return
            
            

            driver.find_element_by_tag_name('body').send_keys(Keys.END)

            sleeping(3)

            driver.find_element_by_tag_name('body').send_keys(Keys.END)

            sleeping(3)

            driver.find_element_by_tag_name('body').send_keys(Keys.HOME)

            sleeping(3)

            tags=driver.find_elements_by_partial_link_text('Contact Supplier')

            for tag in tags:
                
                sleeping(random.randint(min_wait,max_wait))
                
                try:

                    tag.click()

                    sleeping(3)

                    driver.switch_to.window(driver.window_handles[-1])
                    
                    sleeping(2)
                    
                    try:
                        
                        driver.find_element_by_tag_name('body').send_keys(Keys.ESCAPE)
                        
                    except:
                        
                        pass

                    try:

                        company_name=driver.find_element_by_xpath('//span[@class="company-contact"]').text

                        if company_name not in name_of_suppliers:

                            name_of_suppliers.add(company_name)

                            text=driver.find_element_by_tag_name('textarea')

                            text.send_keys(text_message)

                            send_inquiry=driver.find_element_by_xpath('//input[@type="submit"][@value="Send inquiry now"]')

                            send_inquiry.click()
                            
                            count_suppliers_contacted+=1

                            sleeping(3)
        
                        driver.close()

                        driver.switch_to.window(driver.window_handles[0])
                
                        if count_suppliers_contacted>=limit:
                
                            return
            

                    except:

                        pass
                
                except: pass

            next_=next_page()

            if next_:

                send_messages()


window.geometry('400x540')
window.resizable(False,False)
window.title('Bot')

frame_email=Frame(window)
frame_email.grid(row=0,column=0)

label_email=Label(frame_email,text='        Email: ',font=('Sanserif 15 bold'))
label_email.grid(row=0,column=0,pady=10)

email=Entry(frame_email,width=30)
email.grid(row=0,column=1,pady=10)

frame_pwd=Frame(window)
frame_pwd.grid(row=1,column=0)

label_password=Label(frame_pwd,text=' Password: ',font=('Sanserif 15 bold'))
label_password.grid(row=0,column=0,pady=5)

password=Entry(frame_pwd,width=30)
password.grid(row=0,column=1,pady=5)

frame_login=Frame(window)
frame_login.grid(row=2,column=0)

login=Button(frame_login,text='Login',width=13,font=("Sanserif 13 bold"),fg='green',command=login_user)
login.grid(row=0,column=1,pady=10)

frame_wait=Frame(window)
frame_wait.grid(row=3,column=0)

label_min=Label(frame_wait,text='Min Wait: ',font=('Sanserif 13 '))
label_min.grid(row=0,column=0,pady=10)

minn=Entry(frame_wait,width=9)
minn.grid(row=0,column=1,pady=10)

label_max=Label(frame_wait,text='  \tMax Wait: ',font=('Sanserif 13 '))
label_max.grid(row=0,column=2,pady=10)

maxx=Entry(frame_wait,width=9)
maxx.grid(row=0,column=3,pady=10)

frame_message=Frame(window)
frame_message.grid(row=4,column=0)

label_message=Label(frame_message,text='  Message: ',font=('Sanserif 13 '))
label_message.grid(row=0,column=0,pady=20)

message=Text(frame_message,width=35,height=5)
message.grid(row=0,column=1,pady=20)

frame_send_message=Frame(window)
frame_send_message.grid(row=5,column=0)

send_message=Button(frame_send_message,text='Send Messages',font=('Sanserif 12 bold italic'),fg='green',width=17,command=send_messages)
send_message.grid(row=0,pady=5)

frame_update=Frame(window)
frame_update.grid(row=6,column=0,pady=10)

update_label=Label(frame_update,text='')
update_label.grid(row=0,column=0,pady=10)

limit_frame=Frame(window)
limit_frame.grid(row=7,column=0,pady=10)

limit_label=Label(limit_frame,text='Limit number of suppliers:', font=('Sanserif 13'))
limit_label.grid(row=0,column=0)

limit_entry=Entry(limit_frame,width=13)
limit_entry.grid(row=0,column=1)

check_frame=Frame(window)
check_frame.grid(row=8,column=0,pady=5)

check=Checkbutton(check_frame,text="Pause Bot",font=('Sanserif 13'),variable=checked)
check.grid(row=0,column=0)

message_status_frame=Frame(window)
message_status_frame.grid(row=9,column=0,pady=10)

message_status_label=Label(message_status_frame,text='Suppliers contacted: '+str(count_suppliers_contacted),font=('Sanserif 13'))
message_status_label.grid(row=0,column=0)

window.mainloop()
```
